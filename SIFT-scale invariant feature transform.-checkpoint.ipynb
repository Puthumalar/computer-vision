```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read image
img1=cv2.imread("C:/Users/SN111/Pictures/image.jpg")
img1=cv2.resize(img1,(250,250))
img2=cv2.imread("C:/Users/SN111/Pictures/temple.jpg")
img2=cv2.resize(img2,(250,250))
```

```python
#gray scale
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
#SIFT
sift=cv2.SIFT_create()
```

```python
keypoints_1,descriptors_1=sift.detectAndCompute(img1,None)
keypoints_2,descriptors_2=sift.detectAndCompute(img2,None)
```

```python
#feature
bf=cv2.BFMatcher(cv2.NORM_L1,crossCheck=True)
matches=bf.match(descriptors_1,descriptors_2)
matches=sorted(matches,key=lambda x:x.distance)
```

```python
img3=cv2.drawMatches(img1,keypoints_1,img2,keypoints_2,matches[:100],img2,flags=2)
plt.imshow(img3),plt.show()
```
