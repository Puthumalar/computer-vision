# Morphology Transformation and Image Segmentation 

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

# Dilation

```python
img = cv2.imread("C:/Users/MR.H.BALAJEE/Pictures/Screenshots/Screenshot 2024-03-09 112528.png",0)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
_,mask = cv2.threshold(img,220,255,cv2.THRESH_BINARY)
kernal = np.ones((3,3),np.uint8)
dilation = cv2.dilate(mask,kernal,iterations=4)
erode = cv2.erode(mask,kernal,iterations=4)
titles = ["Image","Mask","Dilate","Erode"]
images = [img,mask,dilation,erode]
for i in range(len(images)):
    plt.subplot(2,2,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

# Morphological Opening-errosion following by dilation

```python
open = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernal)
```

```python
cv2.imshow("Original Image",img)
cv2.imshow("Morphological Image",open)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Morphological Closing-errosion following by dilation

```python
close = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernal)
```

```python
cv2.imshow("Original Image",img)
cv2.imshow("Morphological Image",close)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Morphological Gradient - errosion, dialtion, differnce of both

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernal)
```

# Display - Morphological Gradient

```python
cv2.imshow("Original Image",img)
cv2.imshow("Gradient Image",gradient)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Top Hat - opening and diff. of original and opening 

```python
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernal)
```

```python
cv2.imshow("Original Image",img)
cv2.imshow("Tophat Image",tophat)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Black Hat opp. of Tophat

```python
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernal)
```

```python
cv2.imshow("Original Image",img)
cv2.imshow("Black hat Image",blackhat)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
