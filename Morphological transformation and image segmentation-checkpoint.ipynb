```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

# 1.Morphological Dilation


```python
img=cv.imread('C:/Users/SN111/Pictures/P.jfif',0)
img=cv.threshold(img,220,225,cv.THRESH_BINARY)
img
```

```python
img=cv.imread('C:/Users/SN111/Pictures/P.jfif',0)
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
_,mask=cv.threshold(img,220,255,cv.THRESH_BINARY)
titles=['image','mask']
images=[img,mask]
for i in range(len(images)):
    plt.subplot(2,2,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
img=cv.imread('C:/Users/SN111/Pictures/A.webp',0)
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
_,mask=cv.threshold(img,220,255,cv.THRESH_BINARY)
kernal=np.ones((5,5),np.uint8)
dilation=cv.dilate(mask,kernal,iterations=1)
titles=['image','mask','dilate']
images=[img,mask,dilation]
for i in range(len(images)):
    plt.subplot(1,3,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
